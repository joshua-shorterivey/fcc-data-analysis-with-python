You will analyze a dataset of the global average sea level change since 1880. You will use the data to predict the sea level change through year 2050.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

def draw_plot():
    # Read data from file
    df = pd.read_csv('epa-sea-level.csv') 
    x1 = np.linspace(1880, 2050)
    # Create scatter plot
    fig, ax = plt.subplots()
    ax.scatter(x=df.Year, y=df['CSIRO Adjusted Sea Level'])

    # Create first line of best fit
    slope, intercept, r, p, se = linregress(x=df['Year'], y=df['CSIRO Adjusted Sea Level'])
    y1 = slope * x1 + intercept
    ax.plot(x1, y1, color='r')


    # Create second line of best fit
    df_2000 = df[df['Year'] >=2000]
    slope_b, intercept_b, r, p, se = linregress(x=df_2000['Year'], y=df_2000['CSIRO Adjusted Sea Level'])
    y2 = slope_b * x1 + intercept_b
    ax.plot(x1, y2, color='k')

    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Sea Level (Inches)')
    plt.title('Rise In Sea Level')

    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()a
```

```python
draw_plot()
```
